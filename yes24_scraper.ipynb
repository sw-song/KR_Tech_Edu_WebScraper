```python
from selenium import webdriver
chromedriver = '../sw_jupyter/chromedriver'
# selenum의 webdriver에 앞서 설치한 chromedirver를 연동한다.
driver = webdriver.Chrome(chromedriver)
```

```python
home_link= 'http://www.yes24.com/'
links = []
infos = []
for i in range(1,175):
    driver.get('http://www.yes24.com/24/category/bestseller?CategoryNumber=001001003&sumgb=06&PageNumber={}'.format(i))
    links_sele = driver.find_elements_by_css_selector('#bestList > .list > tbody > tr > td.goodsTxtInfo > p > a:nth-of-type(2)')
    for link_sele in links_sele:
        links.append(link_sele.get_attribute('href'))
    
    infos_sele = driver.find_elements_by_css_selector('#bestList > .list > tbody > tr > td.goodsTxtInfo > .aupu')
    for info_sele in infos_sele:
        infos.append(info_sele.text)
    
    print('진행률 = {} / 174'.format(i))
    
```

```python
len(links), len(infos)
```

```python
titles = []
dates = []
prices = []
scores = []
categories = []
sub_categories = []

```

```python
import numpy as np

tempLinks = np.array(links).flatten().tolist()
tempInfos = np.array(infos).flatten().tolist()
```

```python
len(tempLinks), len(tempInfos)
```

```python
count = 0
for i in tempLinks:
    #진행상황 체크
    count += 1
    
    res = requests.get(i)
    html = res.text
    soup = BeautifulSoup(html, 'lxml')

    #카테고리
    temp = soup.select('#infoset_goodsCate a')
    if len(temp) >= 4:
        categories.append(temp[2].text)
        sub_categories.append(temp[3].text)
    elif len(temp) >= 3:
        categories.append(temp[2].text)
        sub_categories.append('일반')
    elif len(temp) >= 2:
        categories.append('일반')
        sub_categories.append('일반')
    #카테고리가 없는 경우(IT 모바일 표기를 찾을 수 없음), 불필요한 페이지로 판단, 다음 링크로 이동 
    else:
        continue

    #도서명(진행상황체크용)
    title = soup.select_one('.gd_titArea > .gd_name').text

    #도서정보
    titles.append(soup.select_one('.gd_titArea > .gd_name').text)
    #dates.append(soup.select_one('.gd_date').text)
    if soup.select_one('span.gd_sellNum > em.divi') != None:
        scores.append(soup.select_one('span.gd_sellNum > em.divi').next_sibling.split()[1])
    else:
        scores.append(0)
    prices.append(soup.select_one('.nor_price > .yes_m').text)

    print('데이터를 저장하고 있습니다. (도서명 : {} | 진행률 : {}/{})'.format(title,count,len(tempLinks)))


```
