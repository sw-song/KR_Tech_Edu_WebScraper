```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
main_link = 'https://www.inflearn.com'
#categories = ['/it-programming','/it','/data-science','/creative','/business','/academics',/%20career']
categories = ['/data-science']
links = []

for category in categories:
    for page in range(1, 7):
        res = requests.get('https://www.inflearn.com/courses{}?order=famous&page={}'\
                          .format(category, page))
        html = res.text
        soup = BeautifulSoup(html, 'lxml')
        temp = soup.select('a.course_card_front')

        # 링크 수집
        links.append([(main_link+x['href']) for x in temp])

        print('링크를 수집중입니다. ({}/45)'.format(page))
```

```python
links[0]
```

```python
titles = []
enrolls = []
prices = []
dates = []
count = 0
for link in links:
    #진행상황 체크
    count += 1
    
    res = requests.get(link)
    html = res.text
    soup = BeautifulSoup(html, 'lxml')

    #카테고리
    title = soup.select_one('.left-box > .description-container > .title')
    titles.append(title.text)
    
    enroll = soup.select_one('.student-cnt')
    enrolls.append(int(enroll.text[:-6]))
    
    if soup.select_one('.del-price')==None:
        price = soup.select_one('.price')
        if price.text == '무료':
            prices.append(0)
        else:
            prices.append(int(price.text[:-1].replace(',','')))
    else:
        price = soup.select_one('.del-price')
        prices.append(int(price.text[2:-1].replace(',','')))
    
    date = soup.select_one('.course_date > div')
    dates.append(date.text.split('(')[0].strip())
    
    print('데이터를 저장하고 있습니다. (강의명 : {} | 진행률 : {}/{})'\
          .format(title.text,count,len(links)))
```

```python
df = pd.DataFrame({'강의명':titles,'등록일':dates,'결제수':enrolls,'가격':prices})
```
