```python
main_link = 'https://www.inflearn.com'
# {카테고리 : 페이지 수} 인덱싱
categories_dict = {'/it-programming':31,'/it':7,'/data-science':8,'/creative':8,'/business':10,'/academics':3,'/career':3, '/life':2}
# 카테고리 명 리스트로 추출
categories_list = list(categories_dict.items())
links = []
count = 0
for category in categories_list: # 카테고리별로 돌아가면서
    for page in range(1, category[1]+1): # 페이지 넘기면서
        res = requests.get('https://www.inflearn.com/courses{}?order=famous&page={}'.format(category[0], page))
        html = res.text # html 추출
        soup = BeautifulSoup(html, 'lxml')
        temp = soup.select('a.course_card_front') # 링크 주소가 있는 태그 추출
        links.append([(main_link+x['href']) for x in temp]) # 리스트에 링크 저장
        count += 1
        print('링크를 수집중입니다. ({}/{})'.format(count,sum(categories_dict.values())))
```

```python
import numpy as np
# 링크 리스트 차원축소(2->1)
tempLinks = sum(links,[])
```

```python
titles = []
enrolls = []
prices = []
dates = []
count = 0
for link in tempLinks: # 링크 리스트 순회
    #진행상황 체크
    count += 1
    
    res = requests.get(link) # 링크 돌아가면서 데이터 요청
    html = res.text # html 추출
    soup = BeautifulSoup(html, 'lxml')

    # 강의명 수집
    title = soup.select_one('#main > section > div.cd-sticky-wrapper > div.cd-header.cd-header__not-owned-course > div > div > div.cd-header__right.ac-cd-7.ac-ct-12 > div.cd-header__title')
    titles.append(title.text)
    
    # 등록수 수집
    if soup.select_one('#main > section > div.cd-sticky-wrapper > div.cd-header.cd-header__not-owned-course > div > div > div.cd-header__right.ac-cd-7.ac-ct-12 > div.cd-header__info-cover > span:nth-of-type(3) > strong')==None:
        enroll = soup.select_one('#main > section > div.cd-sticky-wrapper > div.cd-header.cd-header__not-owned-course > div > div > div.cd-header__right.ac-cd-7.ac-ct-12 > div.cd-header__student_cnt.cd-header__sub-row > span:nth-of-type(2) > strong')
        enrolls.append(enroll.text[:-1])
    else:
        enroll = soup.select_one('#main > section > div.cd-sticky-wrapper > div.cd-header.cd-header__not-owned-course > div > div > div.cd-header__right.ac-cd-7.ac-ct-12 > div.cd-header__info-cover > span:nth-of-type(3) > strong')
        enrolls.append(enroll.text[:-1])
    
    # 가격 수집
    if soup.select_one('.cd-floating__price'):
        price = soup.select_one('.cd-floating__price h4')
        if price.text == '무료':
            prices.append(0)
        elif price.text.find('%') != -1:
            price = soup.select_one('.cd-floating__price h4:nth-of-type(2)')
            prices.append(price)
        else:
            prices.append(price)
        
    elif soup.select_one('.cd-price__pay-price') == None:
        price = soup.select_one('cd-price__reg-price')
        prices.append(price)
            
    else:
        price = soup.select_one('.cd-price__pay-price')
        prices.append(price)
    
    # 오픈일 수집
    date = soup.select_one('#main > section > div.cd-content > div > div > div.ac-cd-8.ac-ct-12 > div > section.cd-date > div > span.cd-date__published-date')
    dates.append(date.text[9:].split('(')[0].strip())
    
    print('데이터를 저장하고 있습니다. (강의명 : {} | 진행률 : {}/{})'\
          .format(title.text,count,len(tempLinks)))
```

```python
df = pd.DataFrame({'강의명':titles,'등록일':dates,'결제수':enrolls,'가격':prices})
```
